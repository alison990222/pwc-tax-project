```python
import pandas as pd
import requests
import json
import time
import random

# 根据页面爬取。淘宝一个页面有44的产品
def format_url(base, pages):    
    urls = []
    for i in range(0, pages*44, 44):
        urls.append(base[:-1] + str(i))
    return urls

# 解析静态html
def parse_page(url,cookies,headers,productCategory):
    result = pd.DataFrame()
    html = requests.get(url,headers = headers,cookies = cookies)
    bs = html.text
    start = bs.find('g_page_config = ') + len('g_page_config = ')
    end = bs.find('"shopcardOff":true}') + len('"shopcardOff":true}')
    js = json.loads(bs[start:end + 1])

    # 嵌套数据所在处
    for i in js['mods']['itemlist']['data']['auctions']:
        product = i['raw_title']  # 产品名
        r = pd.DataFrame({'secondCategory':productCategory,'item':[product]})
        result = pd.concat([result,r])
        
    # 每隔15～30秒发送一次请求
    time.sleep(random.uniform(15.1,30.1))
    return result
```

```python
import jieba as jb
#定义删除除字母,数字，汉字以外的所有符号的函数
def remove_punctuation(line):
    line = str(line)
    if line.strip()=='':
        return ''
    rule = re.compile(u"[^\u4E00-\u9FA5]")
    line = rule.sub('',line)
    return line
 
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords  
 
#加载停用词
stopwords = stopwordslist("chineseStopWords.txt")

#删除除字母,数字，汉字以外的所有符号
result['item'] = result['item'].apply(remove_punctuation)


#分词，并过滤停用词
result['item']= result['item'].apply(lambda x: "".join([w for w in list(jb.cut(x)) if w not in stopwords]))



result= result[["item","secondCategory"]]

      
result
```
